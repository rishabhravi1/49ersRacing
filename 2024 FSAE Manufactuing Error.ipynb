```python
susp = pd.read_excel('2024 FSAE Manufactuing Error.xlsx', sheet_name='Suspension Points')
```

```python
lfpre = susp['Unnamed: 5'][5:17]
lfpost = susp['Unnamed: 7'][5:17]
```

```python
rfpre = susp['Unnamed: 13'][5:17]
rfpost = susp['Unnamed: 15'][5:17]
```

```python
lrpre = susp['Unnamed: 5'][21:36]
lrpost = susp['Unnamed: 7'][21:36]
```

```python
rrpre = susp['Unnamed: 13'][21:36]
rrpost = susp['Unnamed: 15'][21:36]
```

```python
pre = pd.concat([lfpre, rfpre, lrpre, rrpre], axis=0)
post = pd.concat([lfpost, rfpost, lrpost, rrpost], axis=0)
```

```python
len(pre)
```

```python
len(post)
```

```python
import matplotlib.ticker as ticker

sns.histplot(pre, color='blue', label='Pre Weld', binwidth=0.01)
sns.histplot(post, color='red', label='Post Weld', binwidth=0.01)
plt.xlabel('Error in inches')

# Format x-ticks with 3 decimal places
plt.gca().xaxis.set_major_formatter(ticker.FormatStrFormatter('%.3f'))

plt.xticks(np.arange(-.250, 0.250, 0.100))

plt.legend()
plt.title('Pre Weld vs Post Weld')
```
